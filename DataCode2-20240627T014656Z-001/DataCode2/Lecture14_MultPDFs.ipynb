```python
import os
from PyPDF2 import PdfFileReader

def get_info(path):
    with open(path, 'rb') as f:
        pdf = PdfReader(f)
        metadata = pdf.metadata
        num_pages = len(pdf.pages)

        author = metadata.get('/Author', 'Unknown Author')
        creator = metadata.get('/Creator', 'Unknown Creator')
        producer = metadata.get('/Producer', 'Unknown Producer')
        subject = metadata.get('/Subject', 'Unknown Subject')
        title = metadata.get('/Title', 'Unknown Title')

        return {
            'Author': author,
            'Creator': creator,
            'Producer': producer,
            'Subject': subject,
            'Title': title,
            'Number of Pages': num_pages
        }

if __name__ == '__main__':
    #pdf_dir = '/content/drive/MyDrive/LLM_1/ESG/data/ESG_reports'
    pdf_paths = [os.path.join(pdf_dir, filename) for filename in os.listdir(pdf_dir) if filename.lower().endswith('.pdf')]

    for path in pdf_paths:
        pdf_info = get_info(path)
        print(f"PDF: {path}")
        for key, value in pdf_info.items():
            print(f"{key}: {value}")
        print("=" * 30)
```

```python
# Read the PDF file and extract text
pdf_path = '/content/drive/MyDrive/LLM_1/ESG/data/ESG_reports/report_438.pdf'
with open(pdf_path, 'rb') as file:
    pdf_text = ""
    pdf_reader = PyPDF2.PdfReader(file)
    for page in pdf_reader.pages:
        pdf_text += page.extract_text()

# Split text into chunks (you can adjust the chunk size as needed)
chunk_size = 1000
chunks = [pdf_text[i:i + chunk_size] for i in range(0, len(pdf_text), chunk_size)]

# Generate summaries for each chunk
for i, chunk in enumerate(chunks):
    summary = summarizer(chunk, max_length=150, min_length=50, do_sample=False)
    print(f"Summary for Chunk {i + 1}:", summary[0]['summary_text'])
```

```python
# Function to generate summary for a PDF
def generate_summary(pdf_path, chunk_size=1000, max_length=150, min_length=50):
    with open(pdf_path, 'rb') as file:
        pdf_text = ""
        pdf_reader = PyPDF2.PdfReader(file)
        for page in pdf_reader.pages:
            pdf_text += page.extract_text()

        # Split text into chunks
        chunks = [pdf_text[i:i + chunk_size] for i in range(0, len(pdf_text), chunk_size)]

        # Generate summaries for each chunk
        summaries = []
        for i, chunk in enumerate(chunks):
            summary = summarizer(chunk, max_length=max_length, min_length=min_length, do_sample=False)
            summaries.append(summary[0]['summary_text'])

        return "\n".join(summaries)

if __name__ == '__main__':
    pdf_dir = '/content/drive/MyDrive/LLM_1/ESG/data/ESG_reports'

    for filename in os.listdir(pdf_dir):
        if filename.lower().endswith('.pdf'):
            pdf_path = os.path.join(pdf_dir, filename)
            pdf_summary = generate_summary(pdf_path)

            print(f"PDF: {filename}\nSummary:\n{pdf_summary}\n{'=' * 30}")
```
